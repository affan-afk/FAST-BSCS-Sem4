task 1

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
df = pd.DataFrame({
    'sqft': np.random.randint(800, 3000, 40),
    'bedrooms': np.random.randint(1, 6, 40),
    'bathrooms': np.random.randint(1, 4, 40),
    'age': np.random.randint(0, 30, 40),
    'location': np.random.choice(['A','B','C'], 40),
    'price': np.random.randint(100000, 500000, 40)
})
print("sample randomized house data:\n", df.head())

le = LabelEncoder()

df['location'] = le.fit_transform(df['location'])

X = df[['sqft','bedrooms','bathrooms','age','location']]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.4)


model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(f"predicted house prices for sample randomized data: {y_pred}")

print(f"mean squared error: {mean_squared_error(y_test, y_pred)}")
print(f"R^2 Score: {r2_score(y_test, y_pred)}")

new_house = pd.DataFrame([[1600,3,2,5,1]],
columns=['sqft','bedrooms','bathrooms','age','location'])

print(f"new house with features ['sqft','bedrooms','bathrooms','age','location'] = {new_house.iloc[0].tolist()} has predicted price: {model.predict(new_house)[0]}")
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
df = pd.DataFrame({
    'length': np.random.randint(50, 500, 40),
    'links': np.random.randint(0, 10, 40),
    'words_offer': np.random.randint(0, 2, 40),
    'label': np.random.randint(0, 2, 40)
})
X = df[['length','links','words_offer']]
y = df['label']
print("sample email data:\n", df.head())

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


model = SVC(kernel='linear')
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print("predicted classes for test emails:", y_pred)


print("accuracy:",accuracy_score(y_test, y_pred))

new_email = pd.DataFrame([[200,3,1]], columns=['length','links','words_offer'])
print("new email with features ['length','links','words_offer'] = [200, 3, 1] is:",model.predict(new_email))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
df = pd.DataFrame({
    'spending': np.random.randint(100, 3000, 40),
    'age': np.random.randint(18, 60, 40),
    'visits': np.random.randint(1, 30, 40),
    'frequency': np.random.randint(1, 15, 40),
    'label': np.random.randint(0, 2, 40)
})

print("sample randomized Customer data:\n", df.head())


X = df[['spending','age','visits','frequency']]
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.5)


svm = SVC(kernel='linear')
svm.fit(x_train, y_train)

y_pred_svm = svm.predict(x_test)
print("SVM predictions:", y_pred_svm)

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)
print("Decision Tree predictions:", y_pred_dt)


print("SVM accuracy:", accuracy_score(y_test, y_pred_svm))
print("Decision Tree accuracy:", accuracy_score(y_test, y_pred_dt))
```
