```python
#task 1.1
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

class GoalBasedAgent:
    def __init__(self,goal):
        self.goal = goal
    def dls(self,graph, start, depth_limit):
        visited = []
        def dfs(node, depth):
            if depth > depth_limit:
                return None  
            visited.append(node)
            if node == self.goal:
                return (f"Goal found with DLS. Path: {visited}")

            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    path = dfs(neighbor, depth + 1)
                    if path is not None:
                        return path
            visited.pop()  
            return None
        return dfs(start, 0)
class environment:
    def __init__(self,graph):
        self.graph = graph
    def getGraph(self):
        return self.graph

def run_agent(agent,env,start,depth_limit):
    print(agent.dls(env.getGraph(),start,depth_limit))


env = environment(graph)
agent = GoalBasedAgent('I')
run_agent(agent,env,'A',4)
```

```python
#task 1.2
graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}


class utilityBasedAgent:

    def ucs(self,graph, start, goal):
        frontier = [(start, 0)]  
        visited = set()  
        cost_so_far = {start: 0}  
        came_from = {start: None}  

        while frontier:
           
            frontier.sort(key=lambda x: x[1])


           
            current_node, current_cost = frontier.pop(0)


            
            if current_node in visited:
                continue


            
            visited.add(current_node)


            
            if current_node == goal:
                path = []
                while current_node is not None:
                    path.append(current_node)
                    current_node = came_from[current_node]
                path.reverse()
                print(f"Goal found with UCS. Path: {path}, Total Cost: {current_cost}")
                return


            # Explore neighbors
            for neighbor, cost in graph[current_node].items():
                new_cost = current_cost + cost
                if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                    cost_so_far[neighbor] = new_cost
                    came_from[neighbor] = current_node
                    frontier.append((neighbor, new_cost))  

        print("Goal not found")
class environment:
    def __init__(self,graph):
        self.graph = graph
    def getGraph(self):
        return self.graph

def run_agent(agent,env,start,goal):
    agent.ucs(env.getGraph(), start,goal)


env = environment(graph)
agent = utilityBasedAgent()
run_agent(agent,env,'A','I')
```

```python
#task 2
#considering city 1 as starting city the minimum cost will be 80 i.e going on path 1 -> 2 -> 4 -> 3 -> 1 
distances = {
    (1, 2): 10, (2, 1): 10,
    (1, 3): 15, (3, 1): 15,
    (1, 4): 20, (4, 1): 20,
    (2, 3): 35, (3, 2): 35,
    (2, 4): 25, (4, 2): 25,
    (3, 4): 30, (4, 3): 30
}

cities = [2, 3, 4]
min_distance = float(10000000)
best_tour = []

for i in range(len(cities)):
    for j in range(len(cities)):
        if j == i:
            continue
        for k in range(len(cities)):
            if k == i or k == j:
                continue
            tour = [1, cities[i], cities[j], cities[k], 1]
            distance = sum(distances[(tour[m], tour[m+1])] for m in range(len(tour)-1))
            if distance < min_distance:
                min_distance = distance
                best_tour = tour

print("Best tour:", best_tour)
print("Minimum distance:", min_distance)
```

```python
#task 3
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': ['E','I']
}

tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

def dlsWithTree(node, goal, depth, path):
    if depth == 0:
        return False
    if node == goal:
        path.append(node)
        return True
    for child in tree.get(node, []):
        if dlsWithTree(child, goal, depth - 1, path):
            path.append(node)
            return True
    return False

def dlsWithGraph(node, goal, depth, path, visited):
    if depth == 0:
        return False
    if node == goal:
        path.append(node)
        return True
    visited.add(node)
    for child in graph.get(node, []):
        if child not in visited:
            if dlsWithGraph(child, goal, depth - 1, path, visited):
                path.append(node)
                return True
    return False

def iterative_deepeningWithTree(start, goal, max_depth):
    for depth in range(max_depth + 1):
        path = []
        if dlsWithTree(start, goal, depth, path):
            print("Tree Path:", " → ".join(reversed(path)))
            return
    print("Tree: Goal not found")

def iterative_deepeningWithGraph(start, goal, max_depth):
    for depth in range(max_depth + 1):
        path = []
        visited = set()
        if dlsWithGraph(start, goal, depth, path, visited):
            print("Graph Path:", " → ".join(reversed(path)))
            return
    print("Graph: Goal not found")

start_node = 'A'
goal_node = 'I'
max_search_depth = 5

iterative_deepeningWithTree(start_node, goal_node, max_search_depth)
iterative_deepeningWithGraph(start_node, goal_node, max_search_depth)
```
