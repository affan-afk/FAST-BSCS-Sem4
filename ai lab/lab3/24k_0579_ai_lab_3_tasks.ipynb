```python
#task 2
import random
class load_balancer_agent:
    def balance_load(self,servers):
        for i in range(5):
            balanced = False
            if servers[i] == "overloaded":
                for j in range(5):
                    if servers[j] == "underloaded":
                        servers[i] = "balanced"
                        balanced = True
                        print(f"\nserver {i} was balanced with server {j}")
                        break
                if balanced == False:
                    print(f"no server found for server {i} to be balanced with!")
            
class system:
    def __init__(self):
        self.states = ["underloaded","balanced","overloaded"]
        self.servers = []
        for i in range(5):
            self.servers.append(random.choice(self.states))
    def display(self):
        print("CURRENT load OF Servers:")
        for i in range(5):
            print(f"server {i} is {self.servers[i]}")
    def getPercept(self):
        return self.servers
def run_agent(system,agent):
    system.display()
    agent.balance_load(system.servers)
    print()
    system.display()
sys = system()
age = load_balancer_agent()
run_agent(sys,age)
```

```python
#task 3
import random
class backup_agent:
    def retry_backups(self,tasks):
        print()
        for i in range(5):
            while tasks[i] == "Failed":
                if tasks[i] == "Failed":
                    print("Retrying Task: ",i+1,"......")
                    tasks[i] = random.choice(["Completed","Failed"])
                    if tasks[i] == "Failed":
                        print("Task Completion Failed")
                    else:
                        print(f"Task {i+1} Completed Successfully")
            
class server:
    def __init__(self):
        self.status = ["Completed","Failed"]
        self.tasks = []
        for i in range(5):
            self.tasks.append(random.choice(self.status))
    def display(self):
        print("Current Status Of Tasks:")
        for i in range(5):
            print(f"Task {i+1} is {self.tasks[i]}")
    def getPercept(self):
        return self.tasks
def run_agent(server,agent):
    server.display()
    agent.retry_backups(server.getPercept())
    print()
    server.display()
sys = server()
age = backup_agent()
run_agent(sys,age)
```

```python
#task 5
import random

class environment:
    def __init__(self):
        self.corridor = {'A':["ROOM:101","ROOM:102"],'B':["ROOM:103"],'C':["ROOM:104"]} #corridors with rooms
        self.patient_rooms = {"ROOM:101":20230515, "ROOM:102":2323, "ROOM:103":124663, "ROOM:104":292912} #each room has a patient
        self.nurse_stations = {1:"available",2:"busy",3:"busy"}
        self.medicine_storage = {
            "painkillers":["paracetamol","ibuprofen"],
            "antibiotics":["amoxicillin"],
            "thyroid":["levothyroxine"]
        }
        self.schedule = {20230515:"2 pm", 2323:"10 am", 124663:"7 pm", 292912:"5 am"}

    def getPercept_room(self,pID):
        for room, patient in self.patient_rooms.items():
            if pID == patient:
                return room
        return -1

    def getPercept_medicine(self,type):
        return self.medicine_storage[type]

    def getPercept_schedule(self,pID):
        return self.schedule[pID]

    def check_availability(self):
        return random.choice(list(self.nurse_stations.values()))


class robot_agent:
    def __init__(self):
        self.medicine_type = None
        self.currentLocation = "Robot Storage"
        self.goal = None

    def formulate_goal(self,goal):
        self.goal = goal
        print("Goal:", goal)

    def move(self,location):
        self.currentLocation = location
        print("Moved to", location)

    def pickup_medicine(self,environment):
        self.medicine_type = random.choice(["painkillers","antibiotics","thyroid"])
        medicine = random.choice(environment.getPercept_medicine(self.medicine_type))
        print("Picked up", medicine)
        return medicine

    def scan_patient(self,pID):
        print("Scanning patient ID:", pID, "Verified")

    def alert_staff(self):
        print("Alerting nurse station")

    def deliver_medicine(self,environment,pID):
        room = environment.getPercept_room(pID)
        schedule_time = environment.getPercept_schedule(pID)
        current_time = random.choice(["12 am","1 am","2 am","3 am","4 am","5 am","6 am","7 am","8 am","9 am","10 am","11 am","12 pm","1 pm","2 pm","3 pm","4 pm","5 pm","6 pm","7 pm","8 pm","9 pm","10 pm","11 pm"])

        self.move("Medicine Storage")
        medicine = self.pickup_medicine(environment)
        self.move(room)
        self.scan_patient(pID)

        while self.goal != "Medicine Delivered" :
            if current_time == schedule_time:
                self.formulate_goal("Medicine Delivered")
                print("Delivered", medicine, "to room", room)
                break
            else:
                print("Not correct time for medicine, waiting for correct time.....")
            current_time = random.choice(["12 am","1 am","2 am","3 am","4 am","5 am","6 am","7 am","8 am","9 am","10 am","11 am","12 pm","1 pm","2 pm","3 pm","4 pm","5 pm","6 pm","7 pm","8 pm","9 pm","10 pm","11 pm"])

        self.move("Medicine Storage")

        if environment.check_availability() == "busy":
            self.alert_staff()

def run_agent(env,age):
    for patient in env.schedule.keys():
        age.deliver_medicine(env, patient)
        print()
env = environment()
agent = robot_agent()
run_agent(env,agent)
```

```python
#task 6
import random
class Environment:
    def __init__(self):
        self.grid = ['Safe', 'Safe', '🔥',
                    'Safe', '🔥', 'Safe',
                    'Safe', 'Safe', 'Safe']

    def get_percept(self, position):
        return self.grid[position]
    def clean_room(self, position):
        self.grid[position] = 'Safe'
    def display_grid(self, agent_position):
        print("\nCurrent Grid State:")
        grid_with_agent = self.grid[:] 
        grid_with_agent[agent_position] = "🤖" 
        for i in range(0, 9, 3):
            print(" | ".join(grid_with_agent[i:i + 3]))
            print()


class SimpleReflexAgent:
    def __init__ (self):
        self.position = 0 
    def act(self, percept, grid):
        if percept == '🔥':
            grid[self.position] = 'Safe'
            return '🔥 Extinguished'
        else:
            return 'Room is Safe'

    def move(self):
        if self.position < 8:
            self.position += 1
        return self.position
def run_agent(agent, environment, steps):
    for step in range(steps):
        percept = environment.get_percept(agent.position)
        action = agent.act(percept, environment.grid)
        print(f"Step {step + 1}: Position {agent.position} -> Percept - {percept}, Action - {action}")
        environment.display_grid(agent.position)
        if percept == '🔥':
            environment.clean_room(agent.position)
        agent.move()
agent = SimpleReflexAgent()
environment = Environment()

run_agent(agent, environment, 9)
```
