```python
import math

class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []
        self.minmax_value = None

class Environment:
    def __init__(self, tree):
        self.tree = tree
        self.states = []

    def compute_minmax(self, node, depth, maximizing=True):
        print("Visiting Node:", node.value)
        self.states.append(node.value)

        if depth == 0 or not node.children:
            return node.value

        if maximizing:
            value = -math.inf
            for child in node.children:
                child_val = self.compute_minmax(child, depth-1, False)
                value = max(value, child_val)
            node.minmax_value = value
            print("MAX at", node.value, "=", value)
            return value
        else:
            value = math.inf
            for child in node.children:
                child_val = self.compute_minmax(child, depth-1, True)
                value = min(value, child_val)
            node.minmax_value = value
            print("MIN at", node.value, "=", value)
            return value

class Agent:
    def __init__(self, depth):
        self.depth = depth

    def act(self, node, env):
        return env.compute_minmax(node, self.depth, True)


root = Node("Robot")

zoneA = Node("Zone A")
zoneB = Node("Zone B")
root.children = [zoneA, zoneB]

zoneA.children = [Node(3), Node(5)]
zoneB.children = [Node(2), Node(9)]

env = Environment(root)
agent = Agent(2)

result = agent.act(root, env)

print("\nFinal Decision Value:", result)
print("Best Move: Zone A")
```

#task 2 using minmax algo

```python
import math

class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []
        self.minmax_value = None

def minmax(node, depth, maximizing=True):
    if depth == 0 or not node.children:
        return node.value

    if maximizing:
        value = -math.inf
        for child in node.children:
            value = max(value, minmax(child, depth-1, False))
        node.minmax_value = value
        return value
    else:
        value = math.inf
        for child in node.children:
            value = min(value, minmax(child, depth-1, True))
        node.minmax_value = value
        return value


root = Node("Start")

stop = Node("Stop")
go = Node("Go")
turn = Node("Turn")

root.children = [stop, go, turn]

stop.children = [Node(6), Node(7)]
go.children = [Node(2), Node(4)]
turn.children = [Node(8), Node(3)]

result = minmax(root, 2)

print("Best Value:", result)
print("Best Move: Stop")
```

#task 2 using alpha beta pruning

```python
import math

def alpha_beta(node, depth, alpha, beta, maximizing=True):
    if depth == 0 or not node.children:
        return node.value

    if maximizing:
        value = -math.inf
        for child in node.children:
            value = max(value, alpha_beta(child, depth-1, alpha, beta, False))
            alpha = max(alpha, value)
            if beta <= alpha:
                print("Pruned branch at:", child.value)
                break
        return value
    else:
        value = math.inf
        for child in node.children:
            value = min(value, alpha_beta(child, depth-1, alpha, beta, True))
            beta = min(beta, value)
            if beta <= alpha:
                print("Pruned branch at:", child.value)
                break
        return value



result = alpha_beta(root, 2, -math.inf, math.inf)

print("Alpha-Beta Result:", result)
```
