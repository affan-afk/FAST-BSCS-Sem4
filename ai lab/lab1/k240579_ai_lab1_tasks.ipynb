```python
#TASK 3
n = int(input('Enter how many numbers u want to input: '))
numbers = []
even_num = 0
odd_num = 0
for i in range(n):
    numbers.append(int(input(f'Enter Number {i+1}: ')))
    if numbers[i] % 2 == 0:
        even_num+=1
    else:
        odd_num+=1
print('Even Numbers: ',even_num)
print('Odd Numbers: ',odd_num)
print('Numbers Input:')
for i in range(n):
    print(numbers[i])
modifyIndex = int(input('Enter Index to modify: '))
numbers[modifyIndex] = -1
print('Modified List:')
for i in range(n):
    print(numbers[i])
```

```python
#TASK 4
def calculateSalary(salary):
    hra = 0.1*salary
    da =  0.05*salary
    return salary+hra+da
print("Total Salary: $",calculateSalary(5270))
```

```python
#TASK 5
def makeDictionary(myStr):
    u_case = 0
    l_case = 0
    digits = 0
    for i in myStr:
        if i.isupper():
            u_case+=1
        elif i.islower():
            l_case+=1
        elif i.isdigit():
            digits+=1
    myDict = {'String':myStr,'totalChar':len(myStr),'upperCase':u_case,'lowerCase':l_case,'digits':digits}
    return myDict
myDict = makeDictionary(input("Enter String: "))
print(myDict)
```

```python
#TASK 6
myStr = input("Enter String: ")

actual_sub_string = []
for i in range(len(myStr)):
    temp_sub_string = []
    for j in range (i,len(myStr)):
        if (i < len(myStr) - 1 and myStr[i] == myStr[j]):
            temp_sub_string.append(myStr[j])
        else:
            break
    if len(temp_sub_string) > len(actual_sub_string):
        actual_sub_string = temp_sub_string
actual_sub_string = ''.join(actual_sub_string)
print('Longest Substring in ',myStr, ': ',actual_sub_string,' \nLength: ',len(actual_sub_string))
```

```python
#TASK 7
def removeRepetition(numbers):
    new_list = []

    for i in range(len(numbers)):
        add = False
        for j in range(i+1,len(numbers)):
            if numbers[i] == numbers[j]:
                add = True
                for x in range(len(new_list)):
                    if numbers[i] == new_list[x]:
                        add = False
                        break 
                break
        if add == True:
            new_list.append(numbers[i])
    return new_list
numbers = [1,1,1,12,3,4,5,5,5,6,7,5,2,1,1,4,4,5,5]
print(removeRepetition(numbers))
```
