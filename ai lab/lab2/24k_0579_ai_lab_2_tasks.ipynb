```python
#task 1
class SecurityAgent:
    def __init__(self,agent_id, name, status):
        self.agent_id = agent_id
        self.name = name
        self.status = status
class FirewallAgent(SecurityAgent):
    def  monitor_traffic(self):
        print(self.name," is Monitoring Traffic\n")
class MalwareDetectionAgent(SecurityAgent):
    def  scan_files(self):
        print(self.name," is Scanning Files\n")

class AutomationAgent(SecurityAgent):
    def  run_automation(self):
        print(self.name," is Running Automation\n")
agent1 = FirewallAgent(1,"sa","active")
agent2 = MalwareDetectionAgent(2,"ma","active")
agent3 = AutomationAgent(3,"Aa","active")

agent1.monitor_traffic()
agent2.scan_files()
agent3.run_automation()
```

```python
#task 2
class THREAT:
    def __init__(self,threat_id, name, severity):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity
class PhishingThreat(THREAT):
    def  analyze_email(self):
        print("Monitoring Emails for",self.name)
class RansomwareThreat(THREAT):
    def  scan_files(self):
        print("Scanning Files for",self.name)
class BotnetThreat(THREAT):
    def  detect_traffic(self):
        print("Monitoring network traffic for",self.name)

ph = PhishingThreat(1,"spear phishing",2)
rw = RansomwareThreat(2,"Trojans",3)
bn = BotnetThreat(3,"WannaCry",1)

ph.analyze_email()
rw.scan_files()
bn.detect_traffic()
```

```python
#task 3
class ResponseAgent:
    def execute_response(self):
        pass

class AlertAgent(ResponseAgent):
    def execute_response(self):
        print("sending notifications")
class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("blocking malicious activities")
class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("restoring affected systems")
agent1 = AlertAgent()
agent2 = BlockAgent()
agent3 = RecoverAgent()

agent1.execute_response()
agent2.execute_response()
agent3.execute_response()
```

```python
#task 4
class employee:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def work(self):
        return("performed by")
class manager(employee):
    def work(self):
        print("managing",super().work(),self.name)
class Developer(employee):
    def work(self):
        print("developing",super().work(),self.name)
class Designer(employee):
    def work(self):
        print("designing",super().work(),self.name)
obj1 = manager("ali",22)
obj2 = Developer("affan",33)
obj3 = Designer("omer",37)

obj1.work()
obj2.work()
obj3.work()
```

```python
#task 5
class student:
    def __init__(self,name,grade):
        self.__grade = grade
        self.name = name
    def set_grade(self,grade):
        self.__grade = grade
        print("Grade Updated to: ",self.__grade)
    def get_grade(self):
        print("Current Grade: ")
        return(self.__grade)
    def display_info(self):
        print("Student Name: ",self.name)
        print("Student Grade: ",self.__grade)
student1 = student("affan",'A')
print(student1.get_grade())
student1.set_grade('A+')
student1.display_info()
```
